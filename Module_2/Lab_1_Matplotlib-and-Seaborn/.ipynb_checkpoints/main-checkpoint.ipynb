#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
plt.plot(x,y)
plt.title("Plot(x,y)")
plt.xlabel("x dimension")
plt.ylabel("y dimension")
```

```python
plt.plot(x,z)
plt.title("Plot(x,z)")
plt.xlabel("x dimension")
plt.ylabel("z dimension")
```

```python
plt.plot(x, y, label='linear')
plt.plot(x, z, label='quadratic')
plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("plot x,y")

#second axis
plt.subplot(2,2,2)
plt.plot(x,z)
plt.title("plot x,z")
```

```python
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.set_title('LINEAR PLOT')
ax1.plot(x,y,label='linear' )
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')

ax2.set_title('QUAD PLOT')
ax2.plot(x,z, label='quad' )
ax2.set_xlabel('x label')     
ax2.set_ylabel('y label')
fig.legend()

plt.show
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# first axis
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.plot(x,x**2)
ax2.plot(x,np.e**x)
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.set_title('Normal Scale', fontweight='bold')
ax1.plot(x,x**2, "-b", label='same norm')
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')

ax2.set_title('LOG Scale', fontweight='bold')
ax2.plot(x,np.e**x, "-r", label='same log')
ax2.set_xlabel('x label')     
ax2.set_ylabel('y label')

fig.legend(loc="upper center")

plt.show

```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_per_month = fitbit[['Steps','Months_encoded']].groupby('Months_encoded').agg('median')
median_per_month

median_per_month.plot.bar(figsize=(12, 4))
plt.show()
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
median_by_work = fitbit[['Steps','Work_or_Weekend']].groupby('Work_or_Weekend').agg('median')

median_by_work.plot.bar(figsize=(5, 4))
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ['Minutes Sedentary', 'Minutes Lightly Active', 'Minutes Very Active']
for el in features:
    plt.figure()
    sns.scatterplot(x=el, y="Steps", data=fitbit)
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
ax = sns.histplot(data = titanic['Age'])
labels = [f"{stat}: {val:.2f}" for stat, val in titanic['Age'].describe().items()]
handles = [plt.Line2D([], [], visible=False) for _ in labels]

ax.legend(
    handles,
    labels,
)
```
