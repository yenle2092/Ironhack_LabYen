### 1. Import the moovies dataset

### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(df.country,df.movie_title)
```

```python
df1 = pd.crosstab(df.country, df.movie_title,margins=True, margins_name="Total").reset_index()
df1[['country','Total']]
```

```python
df1=pd.crosstab(df['country'], df['color'], margins=True).All
df1
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([df['director_name'], df['genres']], df['country'])
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(df.color,df.color,values=df.duration, aggfunc='mean').round(0)
```

```python
df.groupby('color')['duration'].mean()
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
df['like'] = df['actor_3_facebook_likes']+df['actor_1_facebook_likes']+df['actor_2_facebook_likes']
df.groupby('director_name')['like'].mean()
```
