## 1. Python Create the dataframe

```python
import pandas as pd

data = [[1, 231, "self-employed", "009", "high", 2, 30200],
[2, 765, "students", "005", "high", 3,12700],
[3, 453, "horeca", "007", "medium", 5, 89400],
[4, 231, "self-employed", "009", "high", 2, 30200],
[5, 892, "finance", "003", "low", 3, 740000]]

column_name=["TransactionID", "ClientID", "Profession", "Bank_dep", "Risk", "Number of credits", "Revenue"]

df=pd.DataFrame(data, columns=column_name)
df_label=df
df

```

### Dummy encoding

```python
df_dummy=pd.get_dummies(df)
df_dummy
```

### Label Encoding

```python
from sklearn import preprocessing
label_end=["Profession", "Bank_dep"]
```

```python
label_encoder = preprocessing.LabelEncoder()
df_label[label_end] = df_label[label_end].apply(lambda col: label_encoder.fit_transform(col))
df_label
```

### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
df_ordinal=df
df_ordinal[["Risk"]] = enc.fit_transform(df_ordinal[["Risk"]])
df
```

```python
x = pd.DataFrame(df['Risk'])
x
from sklearn.preprocessing import OrdinalEncoder
ratings_order = ['low', 'medium', 'high']
oe = OrdinalEncoder(categories = [ratings_order], dtype=int)
oe.fit(x)
risk_ord = oe.transform(x)
risk_ord
df['risk_enc']=risk_ord  #new column with the values
```
