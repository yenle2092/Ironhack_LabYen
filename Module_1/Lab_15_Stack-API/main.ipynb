```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
comments = SITE.fetch('comments')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code herefrom stackapi import StackAPI
from datetime import datetime
questions = SITE.fetch('questions',fromdate=datetime(2022,12,1), todate=datetime(2022,12,31))
questions
answers=SITE.fetch('answers',fromdate=datetime(2022,12,1), todate=datetime(2022,12,31))
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
most_voted_question = SITE.fetch('questions',fromdate=datetime(2023,1,26), todate=datetime(2023,1,27), 
                                min=5, tagged='python', sort='votes')
most_voted_question
```
